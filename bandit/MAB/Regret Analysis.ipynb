```python
import pickle
import numpy as np
import pandas as pd
from tqdm import tqdm
import matplotlib.pyplot as plt
from matplotlib.figure import Figure
from IPython.display import display
from typing import List, Dict, Tuple
import os
import gc

PATH = "/Users/sungwoo/ppatteori109@gmail.com - Google Drive/내 드라이브/GSDS/ohlab/Research/latent-context/MAB-results"
```

```python
pkls = dict()
for name in tqdm(os.listdir(PATH)):
    if ".pkl" in name:
        split1 = name.split('.')[0]
        seed = int(split1.split('_')[-1])
        with open(f"{PATH}/{name}", "rb") as f:
            obj = pickle.load(f)
            pkls[seed] = obj

print(len(pkls))
```

```python
for seed in pkls:
    item = pkls[seed]
    for elm in item:
        print(type(elm))
        print(elm.columns)
        display(elm.tail())
    break
```

```python
def regret_process(df:pd.DataFrame) -> Tuple[int, np.ndarray]:
    key_param = df['param'][0]
    nsims = np.sort(df['sim'].unique())
    container = np.zeros(nsims.shape[0], dtype=object)
    for i, sim in enumerate(nsims):
        subdf = df[df['sim'] == sim].copy()
        subdf['instance_regret'] = subdf['optimal_reward'] - subdf['mean_reward']
        container[i] = np.cumsum(subdf['instance_regret'].to_numpy())
    return (key_param, container)
```

```python
regret_results = dict()
for seed in tqdm(pkls):
    regret_dict = dict()
    item = pkls[seed]
    for elm in item:
        key, val = regret_process(elm)
        regret_dict[key] = val
    regret_results[seed] = regret_dict
```

```python
def makeplot(keyword:int, result:dict, label_name=r"$\vert \mathcal{A}_t \vert$", figsize:tuple=(12, 5)) -> Figure:
    NROWS, NCOLS = 1, 2
    fig, (ax1, ax2) = plt.subplots(nrows=NROWS, ncols=NCOLS, figsize=figsize)
    T = 30000
    for key in result:
        item = result[key]
        ax1.plot(np.mean(item, axis=0), label=f"{label_name}={key}")
    ax1.grid(True)
    ax1.set_xlabel("Round")
    ax1.set_ylabel(r"$R_t$")
    ax1.set_title(r"$\bar{R_T}$")
    ax1.legend()
    
    for key in result:
        item = result[key]
        mean = np.mean(item, axis=0)
        std = np.std(item, axis=0, ddof=1)
        ax2.plot(mean, label=f"{label_name}={key}")
        ax2.fill_between(np.arange(T), mean-std, mean+std, alpha=0.2)
    ax2.grid(True)
    ax2.set_xlabel("Round")
    ax2.set_ylabel(r"$R_t$")
    ax2.set_title(r"$\bar{R_T}\pm 1sd$")
    ax2.legend()
    
    fig.tight_layout(rect=[0, 0, 1, 0.95])
    fig.suptitle(f"$R_T$ when seed={keyword}")
    
    plt.show()
    return fig


def save_plot(fig:Figure, path:str, fname:str):
    os.makedirs(path, exist_ok=True)
    fig.savefig(f"{path}/MultiArmedBandit_{fname}.png")
    print("Plot is Saved Completely!")
```

```python
for seed in regret_results:
    print(seed)
    val = regret_results[seed]
    fig = makeplot(keyword=seed, result=val)
    save_plot(fig, path=PATH, fname=f"seed_{seed}")
```
