```python
A = np.random.normal(loc=1, scale=1, size=36).reshape((-1, 6))
print(A)
```

```python
print(np.linalg.det(A))
```

```python
%%time
for _ in range(2000):
    print(np.linalg.inv(A))
```

```python
%%time
A_inv = np.linalg.inv(A)
for _ in range(2000):
    print(A_inv)
```
